```python
from sklearn.ensemble import RandomForestRegressor
from keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np

model2 = RandomForestRegressor(random_state=1, max_depth=10)
lab2 = to_categorical(labels.SQR_catg)
#exDf=exDf.drop('RESULT', axis=1)
model2.fit(df1,lab2)

features2 = df1.columns
importances2 = model2.feature_importances_
indices2 = np.argsort(importances2)
plt.title('Feature Importances')
plt.barh(range(len(indices2)), importances2[indices2], color='b', align='center')
plt.yticks(range(len(indices2)), [features2[i] for i in indices2])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(df11, lab1, test_size=0.3, random_state=1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(df22, lab2, test_size=0.3, random_state=1)
```

```python
from keras.models import Sequential
from keras import layers

input_dim1 = X_train1.shape[1]  # Number of features
modelTF1 = Sequential()
modelTF1.add(layers.Dense(10, input_dim=input_dim1, activation='relu'))
modelTF1.add(layers.Dense(4, activation='sigmoid'))
modelTF1.compile(loss='categorical_crossentropy',
               optimizer='adam',
               metrics=['accuracy'])
history1 = modelTF1.fit(X_train1.to_numpy(), y_train1,validation_data=(X_test1.to_numpy(),y_test1), epochs=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf1 = DecisionTreeClassifier(max_depth=4)

# Train Decision Tree Classifer
clf1 = clf1.fit(X_train1,y_train1)

#Predict the response for test dataset
y_pred1 = clf1.predict(X_test1)
print("Accuracy:",metrics.accuracy_score(y_test1, y_pred1))
```

```python
model1 = RandomForestRegressor(random_state=1, max_depth=10)
model1 = model1.fit(X_train1,y_train1)

#Predict the response for test dataset
y_pred1 = model1.predict(X_test1)
y_pred1 = np.argmax(y_pred1, axis=1)
y_test1 = np.argmax(y_test1, axis=1)
print("Accuracy:",metrics.accuracy_score(y_test1, y_pred1))
```

```python
from sklearn import tree

fn=X_train1.columns
cn=['0','1','2','3']
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (10,2), dpi=900)
for index in range(0, 5):
    tree.plot_tree(model1.estimators_[index],
                   feature_names = fn,
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)
fig.savefig('rf_5trees.png')
```
